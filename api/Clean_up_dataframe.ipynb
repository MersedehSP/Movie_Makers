```python
df.avg_vote = df.avg_vote.astype(float) 
```

```python
print(df.dtypes)
```

```python


###Added new column 'Good_or_Bad_Movie' for avg_vote of 7 or greater = 1 (good) else 0(bad) 
df['Good_or_Bad_Movie'] = np.where(df['avg_vote']>= 7, 1, 0)
```
